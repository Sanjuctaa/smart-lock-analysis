```python
from selenium import webdriver
from time import sleep
from selenium.webdriver.common.by import By

product_name = []
prices = []
rating = []
rating_count = []
review_count = []
link = []

driver = webdriver.Chrome()

for i in range(1, 7):
    driver.get(f"https://www.amazon.in/s?k=smart+lock&page={i}")
    driver.maximize_window()
    sleep(3)  # Wait for the page to load

    all_product = driver.find_elements(By.XPATH, '//*[@data-component-type="s-search-result"]//h2/a')
    
    for p in all_product:
        p.click()
        driver.switch_to.window(driver.window_handles[1])
        sleep(2)

        try:
            title = driver.find_element(By.ID, 'productTitle').text
        except:
            title = 'N/A'
        product_name.append(title)

        try:
            pr = driver.find_element(By.XPATH, '//span[@class="a-price-whole"]').text
        except:
            pr = 'N/A'
        prices.append(pr)

        try:
            rate = driver.find_element(By.XPATH, '/html/body/div[2]/div/div[5]/div[3]/div[4]/div[5]/div/span[1]/span/span[1]/a/span').text
            #<span class="a-size-base a-color-base"> 4.2 </span>
        except:
            rate = 'N/A'
        rating.append(rate)

        try:
            rating_cnt = driver.find_element(By.XPATH, '//span[@id="acrCustomerReviewText"]').text
            #<span id="acrCustomerReviewText" class="a-size-base">8 ratings</span>
        except:
            rating_cnt = 'N/A'
        rating_count.append(rating_cnt)

        try:
            driver.find_element(By.XPATH, '//a[@data-hook="see-all-reviews-link-foot"]').click()
            sleep(2)
            rev_cnt = driver.find_element(By.XPATH, '//div[@data-hook="cr-filter-info-review-rating-count"]').text
        except:
            rev_cnt = 'N/A'
        review_count.append(rev_cnt)

        driver.close()
        driver.switch_to.window(driver.window_handles[0])
        sleep(3)

driver.quit()
```
